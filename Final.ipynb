```python
import pandas as pd

def connect_to_db(database):
    connection = sqlite3.connect(database)
    return connection

# Function to load data from the databases
def load_data():
    # Connect to the first database and load champion data
    connect1 = connect_to_db("lol_draft_data_matchups and synergies.db")
    champion_data = pd.read_sql_query("SELECT * FROM champions", connect1)
    matchups_data = pd.read_sql_query("SELECT * FROM matchups", connect1)
    synergies_data = pd.read_sql_query("SELECT * FROM synergies", connect1)
    connect1.close()
    
    return champion_data, matchups_data, synergies_data
#this function isn't that useful now but I am using it to test my new database
def optimal_pick_for_bot_lane(opponent_champion, champion_data, matchups_data, synergies_data, chosen_role):

    opponent_champion_row = champion_data[champion_data['name'] == opponent_champion]

    if opponent_champion_row.empty:
        return {"error": f"No champion data found for {opponent_champion}"}

    opponent_champion_name = opponent_champion_row.iloc[0]['name']
    opponent_champion_role = opponent_champion_row.iloc[0]['role']


    counter_data = matchups_data[matchups_data['champ2'] == opponent_champion_name]
    if counter_data.empty:
        return {"error": f"No counter data available for {opponent_champion}"}
    
    best_counters = counter_data.sort_values(by='win_rate', ascending=False)

    # Iterate through the sorted counter data to find a valid counter
    for i in range(len(best_counters)):

        top_counter_champion_name = best_counters.iloc[i]['champ1']
        top_counter_win_rate = best_counters.iloc[i]['win_rate']

        # Get the row of the potential counter champion and check the role
        top_counter_champion_row = champion_data[champion_data['name'] == top_counter_champion_name]
        if top_counter_champion_row.empty:
            continue  # Skip if there is some kind of error with the inputted champion/no or not enough data
        top_counter_role = top_counter_champion_row.iloc[0]['role']
        if top_counter_role != chosen_role: #this way you can get a support and an adc
            continue 

        # If a valid counter is found, prepare the recommendation
        top_counter_champion_name = top_counter_champion_row.iloc[0]['name']

        # Get stats of the top counter champion
        counter_champion_stats = top_counter_champion_row.iloc[0]

        recommendation = {
            "champion": top_counter_champion_name,
            "win_rate_against_opponent": top_counter_win_rate,
            "ban_rate": counter_champion_stats['ban_rate'],
            "pick_rate": counter_champion_stats['pick_rate'],
        }
        return recommendation

    # If no suitable counter is found, return an error
    return {"error": "No suitable counter with the same role found."}
def main():
    # Load data from databases
    champion_data, matchups_data, synergies_data = load_data()

    opponent_bot_lane = input(str("Please enter the enemy champion: "))
    role = input(str("Please enter the lane of the champion you want generated: "))

    
    best_pick = optimal_pick_for_bot_lane(opponent_bot_lane, champion_data, matchups_data, synergies_data, role)
    print(best_pick)

if __name__ == "__main__":
    main()
```

```python
def matchup(you, opponent):
    champion_data, matchups_data, synergies_data = load_data()
    matchup_data = matchups_data[
                ((matchups_data['champ1'] == you) & (matchups_data['champ2'] == opponent))]
    win_rate = matchup_data['win_rate'].max()
    #if int(win_rate) < 0 or int(win_rate) > 100:
     #   win_rate = 0
    return win_rate
def synergy(adc,support):
    champion_data, matchups_data, synergies_data = load_data()
    synergy_data = synergies_data[
                ((synergies_data['champ1'] == adc) & (synergies_data['champ2'] == support)) |
                ((synergies_data['champ1'] == support) & (synergies_data['champ2'] == adc))
            ]
    win_rate = synergy_data['win_rate'].max()
   # if int(win_rate) < 0 or int(win_rate) > 100:
    #    win_rate = 0
    return win_rate
def available_duos_to_you(state):
    conn = sqlite3.connect('lol_draft_data_matchups and synergies.db')
    cursor = conn.cursor()
    
    # Retrieve roles from state
    bot = state.get('bot')
    sup = state.get('sup')
    # Get banned champions from state - this is hopefully something I'll integrate later
    banned_champs = set(state.values())
    bot_champions = []
    sup_champions = []
    
    # Fetch champions for bot if not assigned
    if bot is None:
        cursor.execute('''
        SELECT name, win_rate, ban_rate FROM champions
        WHERE role = 'bot' 
        ''')
        bot_champions = cursor.fetchall()
    
    # Fetch champions for sup if not assigned
    if sup is None:
        cursor.execute('''
        SELECT name, win_rate, ban_rate FROM champions
        WHERE role = 'sup'
        ''')
        sup_champions = cursor.fetchall()
    conn.close()
    
    if bot is None and sup is None:
        return [
            (adc[0], support[0], synergy(adc[0],support[0])) for adc in bot_champions 
            for support in sup_champions
            if adc[0] not in banned_champs and support[0] not in banned_champs
        ]
    elif bot is None and sup is not None:
        return [
            (adc[0], sup, synergy(adc[0], sup)) for adc in bot_champions 
            if adc[0] not in banned_champs
        ]
    elif bot is not None and sup is None:
        return [
            (bot, support[0], synergy(bot, support[0])) for support in sup_champions 
            if support[0] not in banned_champs
        ]
    else:
        return [(bot, sup, synergy(bot,sup))]
    #this function will tell me all the available supports and adc's and can be called based on the available champs

def available_duos_for_opponent(state):
    conn = sqlite3.connect('lol_draft_data_matchups and synergies.db')
    cursor = conn.cursor()
    
    # Retrieve roles from state
    bot = state.get('opp_bot')
    sup = state.get('opp_sup')
    # Get banned champions from state
    banned_champs = set(state.values())
    bot_champions = []
    sup_champions = []
    
    # Fetch champions for bot if not assigned
    if bot is None:
        cursor.execute('''
        SELECT name, win_rate, ban_rate FROM champions
        WHERE role = 'bot' 
        ''')
        bot_champions = cursor.fetchall()
    
    # Fetch champions for sup if not assigned
    if sup is None:
        cursor.execute('''
        SELECT name, win_rate, ban_rate FROM champions
        WHERE role = 'sup'
        ''')
        sup_champions = cursor.fetchall()
    conn.close()
    if bot is None and sup is None:
        return [
            (adc[0], support[0], synergy(adc[0],support[0])) for adc in bot_champions 
            for support in sup_champions
            if adc[0] not in banned_champs and support[0] not in banned_champs
        ]
    elif bot is None and sup is not None:
        return [
            (adc[0], sup, synergy(adc[0], sup)) for adc in bot_champions 
            if adc[0] not in banned_champs
        ]
    elif bot is not None and sup is None:
        return [
            (bot, support[0], synergy(bot, support[0])) for support in sup_champions 
            if support[0] not in banned_champs
        ]
    else:
        return [(bot, sup, synergy(bot,sup))]
#same function as above, but for the enemy team
```

```python
#minmax for the three solo lanes
def min_max_solo(state, role=None):
    champion_data, matchups_data, synergies_data = load_data()
    opp = state.get('opp')

    #if opponent has picked a champion, find the best counter
    if opp:
        counters = matchups_data[matchups_data["champ2"] == opp]
        if counters.empty:
            return f"No counter data found for {opp}"
        
        best_counter = counters.loc[counters["win_rate"].idxmin()]["champ1"]
        best_counter_wr = counters["win_rate"].min()

        return f"Best counter to {opp}: {best_counter} (Opponent win rate: {best_counter_wr}%)"

    #If no opponent pick, find the champion with the best worst matchup
    elif role:
        available_champs = champion_data[champion_data["role"] == role]
        if available_champs.empty:
            return f"No champions found for role {role}"

        best_champ = None
        best_worst_wr = 0 

        for champ in available_champs["name"]:
            matchups_as_champ1 = 100 - matchups_data[matchups_data["champ1"] == champ]["win_rate"]
            matchups_as_champ2 = matchups_data[matchups_data["champ2"] == champ]["win_rate"]# Reverse the win rate
            #combine all win rates for this champion
            all_win_rates = matchups_as_champ1.tolist() + matchups_as_champ2.tolist()

            worst_wr = min(all_win_rates)  #sort for worst matchup for the champion
            if worst_wr > best_worst_wr:
                best_worst_wr = worst_wr
                best_champ = champ

    return f"Best overall pick for {role}: {best_champ} (Worst matchup: {best_worst_wr}%)"



state1 = {'you': None, 'opp': "Jax"}  # Find best counter to Jax
state2 = {'you': None, 'opp': None}  #pick best option for user with minmax

print(min_max_solo(state1))
print(min_max_solo(state2, role="top"))
```

```python
import itertools
import pandas as pd

# Load data from the database
champion_data, matchups_data, synergies_data = load_data()

# Create lists of ADCs and Supports
adc = champion_data[champion_data["role"] == "bot"]["name"].tolist()
support = champion_data[champion_data["role"] == "sup"]["name"].tolist()

# Generate all ADC-Support pairs
bot_lane_combos = list(itertools.product(adc, support))

# Generate valid 2v2 matchups (no repeated champions)
valid_matchups = []
for team_a, team_b in itertools.product(bot_lane_combos, repeat=2):
    adc_a, sup_a = team_a
    adc_b, sup_b = team_b

    # Ensure there are no duplicate champions across both teams
    if len({adc_a, sup_a, adc_b, sup_b}) == 4:  
        valid_matchups.append((team_a, team_b))

# Create a DataFrame to store the matchups
df = pd.DataFrame(valid_matchups, columns=["Team A", "Team B"])
df["Value"] = None

# Compute matchup values using min_max() and normalize them
for index, (team_a, team_b) in enumerate(valid_matchups):
    adc_a, sup_a = team_a
    adc_b, sup_b = team_b

    # Define state for original matchup
    state_original = {
        'sup': sup_a, 
        'bot': adc_a,                  
        'opp_bot': adc_b,        
        'opp_sup': sup_b            
    }

    # Define state for reverse matchup
    state_reverse = {
        'sup': sup_b, 
        'bot': adc_b,                  
        'opp_bot': adc_a,        
        'opp_sup': sup_a            
    }

    # Calculate both matchup values
    original_score = min_max(state_original, 'min')
    reverse_score = min_max(state_reverse, 'min')

    # Avoid division by zero
    total_score = original_score + reverse_score
    normalized_score = original_score / total_score if total_score > 0 else 0

    # Store the normalized value
    df.at[index, "Value"] = normalized_score

    # Print in a readable format
    print(f"{team_a} vs {team_b}  Original: {original_score:.2f}, Reverse: {reverse_score:.2f}, Normalized: {normalized_score:.4f}")

# Save to CSV
df.to_csv("bot_lane_2v2_table.csv", index=False)

print("Bot lane 2v2 table saved successfully!")
```

```python
import pandas as pd
import ast

# Function to compute worst-case (max) win rates for 1v1 from 2v1 files
def compute_1v1_max(input_file, champion_col, opponent_col, matchup, source):
    df = pd.read_csv(input_file)

    # Convert string representations of tuples into actual tuples
    df[champion_col] = df[champion_col].apply(ast.literal_eval)

    # Create lists to store separate matchups
    extracted_matchups = []

    # Process each row, splitting the key column into separate ADC and Support matchups
    for (champion, support), opponent, value in df.itertuples(index=False):
        if matchup == "adc_vs_adc":
            extracted_matchups.append((champion, opponent, value))  # Max winrate for ADC vs ADC matchups
        elif matchup == "adc_vs_sup":
            extracted_matchups.append((champion, opponent, value))  # Max winrate for ADC vs Support matchups
        elif matchup == "sup_vs_adc":
            extracted_matchups.append((support, opponent, value))  # Max winrate for Support vs ADC matchups
        elif matchup == "sup_vs_sup":
            extracted_matchups.append((support, opponent, value))  # Max winrate for Support vs Support matchups

    # Create DataFrame from extracted matchups
    df_extracted = pd.DataFrame(extracted_matchups, columns=["Champion", "Opponent", "Value"])

    # Get the best-case (maximum) win rate per champion-opponent pair
    df_max = df_extracted.groupby(["Champion", "Opponent"])["Value"].max().reset_index()

    # Rename columns for clarity
    df_max.columns = ["Champion", "Opponent", "Winrate"]

    # Save to CSV with formatted filename
    formatted_filename = f"1v1case_{matchup}_({source}).csv"
    df_max.to_csv(formatted_filename, index=False)
    print(f"{formatted_filename} saved successfully!")

# Generate 1v1 files (Maximizing from the 2v1 files)
compute_1v1_max("2v1case_adc_sup_vs_adc.csv", "Key", "Opponent", "adc_vs_adc", "adc_sup_vs_adc")  # ADC vs ADC from adc_sup_vs_adc
compute_1v1_max("2v1case_adc_sup_vs_sup.csv", "Key", "Opponent", "adc_vs_sup", "adc_sup_vs_sup")  # ADC vs Sup from adc_sup_vs_sup
compute_1v1_max("2v1case_adc_sup_vs_adc.csv", "Key", "Opponent", "sup_vs_adc", "adc_sup_vs_adc")  # Sup vs ADC from adc_sup_vs_adc
compute_1v1_max("2v1case_adc_sup_vs_sup.csv", "Key", "Opponent", "sup_vs_sup", "adc_sup_vs_sup")  # Sup vs Sup from adc_sup_vs_sup
```
