```python
import sqlite3
import pandas as pd

def connect_to_db(database):
    connection = sqlite3.connect(database)
    return connection

# Function to load data from the databases
def load_data():
    # Connect to the first database and load champion data
    connect1 = connect_to_db("lol_draft_data_worlds2023.db")
    champion_data = pd.read_sql_query("SELECT * FROM champions", connect1)
    connect1.close()

    # Connect to the second database and load matchups and synergies data
    connect2 = connect_to_db("lol_draft_data_worlds2023_counters_and_synergies.db")
    matchups_data = pd.read_sql_query("SELECT * FROM matchups", connect2)
    synergies_data = pd.read_sql_query("SELECT * FROM synergies", connect2)
    connect2.close()

    return champion_data, matchups_data, synergies_data

def optimal_pick_for_bot_lane(opponent_champion, champion_data, matchups_data, synergies_data, chosen_role):

    opponent_champion_row = champion_data[champion_data['name'] == opponent_champion]

    if opponent_champion_row.empty:
        return {"error": f"No champion data found for {opponent_champion}"}

    opponent_champion_name = opponent_champion_row.iloc[0]['name']
    opponent_champion_role = opponent_champion_row.iloc[0]['role']


    counter_data = matchups_data[matchups_data['champ2'] == opponent_champion_name]
    if counter_data.empty:
        return {"error": f"No counter data available for {opponent_champion}"}
    
    best_counters = counter_data.sort_values(by='win_rate', ascending=False)

    # Iterate through the sorted counter data to find a valid counter
    for i in range(len(best_counters)):
        # Skip if the number of games is not significant
        if best_counters.iloc[i]['no_games'] <= 1:
            continue

        top_counter_champion_name = best_counters.iloc[i]['champ1']
        top_counter_win_rate = best_counters.iloc[i]['win_rate']
        top_counter_no_games = best_counters.iloc[i]['no_games']

        # Get the row of the potential counter champion and check the role
        top_counter_champion_row = champion_data[champion_data['name'] == top_counter_champion_name]
        if top_counter_champion_row.empty:
            continue  # Skip if there is some kind of error with the inputted champion/no or not enough data
        top_counter_role = top_counter_champion_row.iloc[0]['role']
        if top_counter_role != chosen_role: #this way you can get a support and an adc
            continue 

        # If a valid counter is found, prepare the recommendation
        top_counter_champion_name = top_counter_champion_row.iloc[0]['name']

        # Get stats of the top counter champion
        counter_champion_stats = top_counter_champion_row.iloc[0]

        recommendation = {
            "champion": top_counter_champion_name,
            "win_rate_against_opponent": top_counter_win_rate,
            "no_of_games_against_opponent": top_counter_no_games,
            "ban_rate": counter_champion_stats['ban_rate'],
            "presence": counter_champion_stats['presence'],
            # "synergy_win_rate": synergy_win_rate
        }
        return recommendation

    # If no suitable counter is found, return an error
    return {"error": "No suitable counter with the same role found."}
def main():
    # Load data from databases
    champion_data, matchups_data, synergies_data = load_data()

    opponent_bot_lane = input(str("Please enter the enemy champion: "))
    role = input(str("Please enter the lane of the champion you want generated: "))

    
    best_pick = optimal_pick_for_bot_lane(opponent_bot_lane, champion_data, matchups_data, synergies_data, role)
    print("Optimal Pick Recommendation:")
    print(f"Champion: {best_pick['champion']}")
    print(f"Win Rate against opponent: {best_pick['win_rate_against_opponent']:.2f}")
    print(f"Number of games against opponent: {best_pick['no_of_games_against_opponent']}")
    print(f"Ban Rate: {best_pick['ban_rate']:.2f}")
    print(f"Presence: {best_pick['presence']:.2f}")

if __name__ == "__main__":
    main()
```

```python
champion_data, matchups_data, synergies_data = load_data()
def matchup(you, opponent):
    matchup_data = matchups_data[
                ((matchups_data['champ1'] == you) & (matchups_data['champ2'] == opponent))]
    win_rate = matchup_data['win_rate'].max()
    #if int(win_rate) < 0 or int(win_rate) > 100:
     #   win_rate = 0
    return win_rate
def optimal_picks_for_bot_lane(opponent_champion, champion_data, matchups_data, synergies_data, chosen_role1, chosen_role2):
    bot_champ_stats = []
    sup_champ_stats = []
    
    opponent_champion_row = champion_data[champion_data['name'] == opponent_champion]
    if opponent_champion_row.empty:
        return {"error": f"No champion data found for {opponent_champion}"}

    opponent_champion_name = opponent_champion_row.iloc[0]['name']
    opponent_champion_role = opponent_champion_row.iloc[0]['role']

    counter_data = matchups_data[matchups_data['champ2'] == opponent_champion_name]
    if counter_data.empty:
        return {"error": f"No counter data available for {opponent_champion}"}
    
    best_counters = counter_data.sort_values(by='win_rate', ascending=False)

    # Get best ADC counters
    for i in range(len(best_counters)):
        if best_counters.iloc[i]['no_games'] <= 1:
            continue

        top_counter_champion_name = best_counters.iloc[i]['champ1']
        top_counter_win_rate = best_counters.iloc[i]['win_rate']
        top_counter_no_games = best_counters.iloc[i]['no_games']

        # Get the row of the potential counter champion and check the role
        top_counter_champion_row = champion_data[champion_data['name'] == top_counter_champion_name]
        if top_counter_champion_row.empty:
            continue

        top_counter_role = top_counter_champion_row.iloc[0]['role']
        if top_counter_role != chosen_role1:  # Ensures this is the ADC/support role
            continue

        bot_champ_stats.append(top_counter_champion_row.iloc[0])

    # Get best Support counters
    for i in range(len(best_counters)):
        if best_counters.iloc[i]['no_games'] <= 1:
            continue

        top_counter_champion_name = best_counters.iloc[i]['champ1']
        top_counter_win_rate = best_counters.iloc[i]['win_rate']
        top_counter_no_games = best_counters.iloc[i]['no_games']

        top_counter_champion_row = champion_data[champion_data['name'] == top_counter_champion_name]
        if top_counter_champion_row.empty:
            continue

        top_counter_role = top_counter_champion_row.iloc[0]['role']
        if top_counter_role != chosen_role2:  # Ensures this is the Support role
            continue

        sup_champ_stats.append(top_counter_champion_row.iloc[0])

    # Find optimal ADC-Support synergy
    best_synergy_score = 0
    best_pair = None

    for bot in bot_champ_stats:
        for sup in sup_champ_stats:
            # Find synergy between bot and sup based on champ IDs
            synergy_data = synergies_data[
                ((synergies_data['champ1'] == bot['name']) & (synergies_data['champ2'] == sup['name'])) |
                ((synergies_data['champ1'] == sup['name']) & (synergies_data['champ2'] == bot['name']))
            ]

            if synergy_data.empty:
                continue  # Skip if no synergy data is found for this pair

            # Calculate the best synergy win rate for this ADC-Support pair
            synergy_win_rate = synergy_data['win_rate'].max()

            # If this synergy is the best we've seen, update our best pair
            if synergy_win_rate > best_synergy_score:
                best_synergy_score = synergy_win_rate
                best_pair = {
                    "adc": bot['name'],
                    "support": sup['name'],
                    "synergy_win_rate": best_synergy_score,
                    "adc_win_rate": matchup(bot['name'], opponent_champion_name),
                    "sup_win_rate": matchup(sup['name'], opponent_champion_name)
                }

    # Return the best ADC and Support pair based on synergy
    if best_pair:
        return best_pair
    else:
        return {"error": "No suitable synergy data found for any ADC-Support pair"}

    

    
    
def main():
    # Load data from databases
    champion_data, matchups_data, synergies_data = load_data()

    opponent_bot_lane = input(str("Please enter the enemy bot lane: "))

    # Get the best pick recommendation
    best_pick = optimal_picks_for_bot_lane(opponent_bot_lane, champion_data, matchups_data, synergies_data, 'bot', 'sup')
    print(best_pick)
    
# Run the function
if __name__ == "__main__":
    main()
```
